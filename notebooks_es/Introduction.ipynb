## Fundamentos de la informática (CS)
- el núcleo es la capacidad disciplinada de ser lógico y creativo de una manera pragmática para resolver problemas en diversas disciplinas
- La informática es una disciplina más nueva que surge de las matemáticas, la ingeniería y las ciencias naturales.
- Al igual que los matemáticos, los informáticos utilizan lenguajes formales para denotar ideas (especialmente la computación).
- Al igual que los ingenieros, diseñan cosas, ensamblan componentes en sistemas y evalúan compensaciones entre alternativas.
- Al igual que los científicos, observan el comportamiento de sistemas complejos, formulan hipótesis y prueban predicciones.

- **la habilidad más importante para un informático es la resolución de problemas, principalmente escribir programas de computadora**
- el objetivo de este curso es enseñarte a pensar como un informático
- El trabajo principal de los informáticos gira en torno a problemas, algoritmos y programas.

## Problemas, Algoritmos y Programas

### Problema
- Nos ocupamos y resolvemos muchos problemas en todos los ámbitos de la vida.
- Un problema es una pregunta planteada para una investigación que alguien necesita responder o encontrar una solución.
- Los informáticos suelen ocuparse de problemas computacionales.
- puede ser tan simple como: 
    - ¿Cuál es la suma de 9 y 999?
- o puede ser tan complicado como:
    - ¿Cuál es el camino más corto desde San Francisco, California a la ciudad de Nueva York, Nueva York?
- uno debe comprender el problema, analizar los requisitos, limitaciones y suposiciones para resolverlo correctamente.

### Algoritmo
- una vez que el problema está formulado y bien analizado, los informáticos trabajan en algoritmos
- proceso/tarea paso a paso para resolver un problema determinado
    - como una receta para un menú de comida
- típicamente escrito en lenguaje humano o pseudocódigo (en el medio)
- p.ej. Problema: ¿Cómo puede tu amigo marciano comprar comestibles en la Tierra?
- Deberías poder ayudar a resolver este problema dado que vives en la Tierra y compras comestibles muchas veces.
    - en otras palabras, eres el experto en el dominio
- pasos del algoritmo:
    1.Haz una lista de compras
    2. Conduce hasta una tienda de comestibles.
    3. Estaciona tu auto
    4. Encuentra elementos en la lista.
    5. Pagar
    6. Cargar comestibles
    7. Conducir a casa
- faltan muchos detalles en estos pasos
    - es un buen comienzo y se puede perfeccionar profundizando cada paso más hacia abajo

### Programa
- Una vez finalizados los pasos del algoritmo, los programadores pueden convertirlos en instrucciones de computadora.
- secuencia de instrucciones que especifica cómo realizar un cálculo usando computadoras
    - el cálculo puede ser matemático (resolución de sistemas de ecuaciones), cálculo simbólico (búsqueda y sustitución de texto en un documento, simulaciones científicas), etc.
- las instrucciones (o comandos o declaraciones) se ven diferentes en diferentes lenguajes de programación, pero los conceptos básicos fundamentales son los mismos
- algunos conceptos fundamentales que conforman un programa de computadora independientemente del idioma son:
    1. datos/valores y variables
    2. entrada
    3. salida
    4. matemáticas
    5. ejecución condicional (lógica)
    6. repetición

![](activos/BuildingBlocks.png)

### datos y variables
- el programa trabaja con datos (llamados valores) que deben almacenarse en la memoria del ordenador; Las variables son nombres dados a ubicaciones de memoria para almacenar valores.

### entrada
- obtener datos del teclado, un archivo o algún dispositivo

### salida
- mostrar datos/respuestas en la pantalla, o guardarlos en un archivo o en un dispositivo

### matemáticas
- operaciones matemáticas básicas como suma, resta, multiplicación, etc.

### condicionales
- probar ciertas condiciones o lógicas y ejecutar la secuencia apropiada de declaraciones

### bucles
- realizar alguna acción repetidamente, generalmente con alguna variación cada vez

## Lenguajes de programación
- El lenguaje de programación es un lenguaje formal utilizado para crear programas de computadora.
- hay docenas de lenguajes de programación

### Tipos de lenguajes de programación
### Idiomas de alto nivel
- lenguajes diseñados para ser amigables para los programadores ocultando todos los detalles
    - C++, Java, C, FORTRAN, Python, PhP, JavaScript, Rust, etc.
- ventajas: 
    - más simple; más fácil de aprender y escribir
    - más corto y más fácil de leer
    - los programas son portátiles; Puede ejecutarse en diferentes máquinas con algunas o ninguna modificación.
- desventajas:
    - la traducción a código de máquina puede llevar algún tiempo
    - más lento de ejecutar si la traducción no es óptima


### Idiomas de bajo nivel
- lenguaje de máquina, por ejemplo, lenguaje ensamblador
- En términos generales: las computadoras sólo pueden ejecutar programas escritos en lenguajes de bajo nivel.
- los programas escritos en idiomas de alto nivel deben traducirse antes de poder ejecutarse
- ventaja: 
    - los programas se ejecutan más rápido
- desventajas:
    - más difícil de aprender y escribir código (es necesario conocer detalles de muy bajo nivel sobre las computadoras)
    - los programas no son portátiles; Generalmente es necesario reescribir para cada tipo de arquitectura de máquina.
 

## Formas de traducir programas de alto nivel.
- Hay dos formas de traducir programas de alto nivel: **interpretar** y **compilar**

### interpretando
- un intérprete lee un programa de alto nivel y hace lo que dice
- traduce el programa línea por línea, alternativamente leyendo líneas y ejecutando comandos
- Python, PhP, JavaScript son lenguajes interpretados.

![](recursos/Intrepreter.png)

### compilando
- un compilador lee un programa de alto nivel y lo traduce todo de una vez en código de bytes antes de ejecutar cualquiera de los comandos
- los compiladores verifican la sintaxis/gramáticas de los idiomas
- el código de bytes o el programa binario debe cargarse en la memoria para ejecutarse
- C++, C, Rust, Java, FORTRAN son lenguajes de programación compilados.

![](recursos/Compiler.png)

## Lenguaje de programación C++
- C++ es uno de los lenguajes de programación de uso general más populares; consulte [índice de tiobe](https://www.tiobe.com/tiobe-index/)
- lenguaje compilado de alto nivel
- extensión del lenguaje de programación C
    - misma sintaxis; Excava todas las bibliotecas C y admite clases (programación orientada a objetos, POO)
- puedes usar todas las bibliotecas y funciones de C en C++
- Diseñado para programación de sistemas y software integrado con recursos limitados y sistemas grandes con rendimiento, eficiencia y flexibilidad.
- consulte la entrada de Wikipedia para conocer la historia y otros detalles: - https://en.wikipedia.org/wiki/C%2B%2B
- sitio oficial de referencia de C++ https://en.cppreference.com/w/

## Resolución de problemas usando C++
- C++ es una de las muchas herramientas que utilizan los informáticos para resolver problemas.
- aprender y dominar el uso de sus herramientas es importante
- El objetivo final es aprender a resolver problemas como científicos informáticos.
    - escribir código, aunque importante, es una pequeña parte de la resolución de problemas
    - Los programadores dedican aproximadamente el 20% de su tiempo a desarrollar código.
- el diseño y desarrollo del programa requiere muchos pasos 
    - una parte importante del proceso de ingeniería de software
    
### Pasos de desarrollo de C++
- En la siguiente figura se muestra un enfoque simple de alto nivel para desarrollar programas C++.
![](recursos/CPPDevelopment.png)

#### Paso 1: Definir el problema
- descubre qué problema estás tratando de resolver
- puede ser sólo una idea o el planteamiento de un problema completamente investigado
    1. escribe un programa para encontrar las temperaturas promedio en los EE. UU. durante la última década
    - escribir un programa que compruebe si una cadena determinada es un palíndromo
    - escribir un programa que encuentre el camino más corto desde la ciudad de Nueva York a San Francisco
    - diseñar y desarrollar un sistema para el rover de Marte
    
#### Paso 2: Analizar el problema
- comprender realmente el alcance y todos los parámetros del problema
- recopilar todos los requisitos, describir las suposiciones, restricciones de entrada y salida, etc.
    - digamos que desea resolver el problema de temperatura n.° 1 definido en el paso 1
    - ¿Cómo obtuviste los datos de temperaturas? ¿Incluirías los 50 estados?
    - ¿Qué pasa si no hay datos de temperatura para algunos estados? ¿Vas a promediar los estados y luego encontrar el promedio del promedio?
    - ¿Hay valores atípicos? ¿Cómo los manejaste?

#### Paso 3: Solución de diseño
- determinar "cómo" resolverías realmente el problema
- muchas maneras de resolver un problema; examinar las compensaciones, p. eficiencia versus costo, etc.
    - a menudo las soluciones simples y directas son mejores
- dividir el problema en módulos más pequeños o subproblemas
    - escribir pasos de algoritmo para cada subproblema
    - la solución modular es más fácil de actualizar, ampliar y reutilizar sin afectar otras partes
- diseñar maquetas; diseño del sistema de dibujo
    - explicar cómo interactúan varios módulos y componentes entre sí
    - ayuda a abordar cualquier suposición hecha o limitación

#### Paso 4: Implementar la solución
- escribir el programa usando un lenguaje de programación 
    - utilizar C++ en este curso
- Los programadores dedican sólo alrededor del 20% de su tiempo a escribir código.
- necesita una computadora con un editor de texto o un entorno de desarrollo integrado (IDE)
    - depende del sistema: Windows, Linux, Mac, etc.
- un buen editor de código o IDE normalmente proporciona
    - forma de organizar proyectos con múltiples archivos y recursos
    - resaltado de sintaxis, codificación de colores, números de línea, forma sencilla de compilar, ejecutar y depurar código

#### Paso 5: Crear programa
- Este es normalmente un proceso de dos pasos:
    Primero, compilar código C ++ en código objeto o byte
        - un proyecto/programa puede contener muchos archivos c++ y archivos de encabezado (extensiones de archivo .cpp, .cc, .h)
        - cada archivo fuente C++ se convierte en un archivo objeto (normalmente tiene extensión .o)
    2. vincular archivos de objetos y bibliotecas
        - el programa llamado enlazador combina o vincula todos los archivos objeto y la biblioteca estándar de C++ y cualquier otra biblioteca utilizada en un solo programa ejecutable
- los compiladores modernos (por ejemplo, g++) pueden realizar ambos pasos a la vez
- **Makefile** es una mejor manera de crear programas C/C++
    - un script tipo bash que simplifica muchos pasos para crear programas una y otra vez
    - aprenda sobre makefile con estos tutoriales:
        - https://makefiletutorial.com/
        - https://www.cs.bu.edu/teaching/cpp/writing-makefiles/

#### Paso 6, 7 y 8: ejecutar, probar y depurar el programa
- debes ejecutar o ejecutar el programa para probarlo
- un programa llamado cargador carga el ejecutable en la memoria principal RAM (Random Access Memory)
    - CPU (Unidad Central de Procesamiento) hace el cálculo real
- Las pruebas garantizan que obtendrá resultados correctos bajo todos los supuestos.
- Los programadores pueden pasar mucho tiempo probando sus propios programas o los de otros.
    - Contrariamente a la intuición, ¡intentas romper tu propio programa!
- si existe un error o error, debe señalarlo y corregirlo
    - construir el programa nuevamente repitiendo desde el paso 5 tantas veces como sea necesario
- dos formas comunes de probar la corrección de su programa; ¡Aprende ambos en este curso!
    1. ejecutar y alimentar manualmente los datos de entrada y comparar los resultados con las respuestas esperadas
    - escriba casos de prueba y pruebe su programa automáticamente usando código
    

#### Paso 9: Implementar el programa
- entregar o implementar su programa en un entorno de producción
    - dado que su programa cumple con todos los requisitos, pasa pruebas rigurosas, etc.

## Estructura de un programa C++
- El programa C++ puede constituir uno o varios archivos de texto plano. 
    - normalmente archivos de encabezado y fuente
- cada archivo C++ contiene varias declaraciones, instrucciones y códigos C++
- El archivo fuente C++ normalmente tiene estas extensiones: `filename.cpp` o `filename.cc`
    - evitar espacios en los nombres de archivos y carpetas
- El programa C++ debe tener un archivo con la función **main()**
    - **int main()** es la entrada principal del programa
    - la computadora comienza a ejecutar instrucciones de arriba a abajo comenzando desde **main()**
    
- El archivo C++ normalmente contiene:
    - **descripción del programa**
        - breve información sobre el programa y el programador, información sobre derechos de autor
        - estos son comentarios destinados a programadores/lectores
        
    - **bibliotecas**
        - incluir los librabires (archivos de encabezado) que solo son necesarios
        - Las bibliotecas proporcionan códigos integrados que los programadores pueden usar.
        - los programadores no tienen que escribir todos los detalles básicos y las tareas comunes
            - para que puedan concentrarse en resolver problemas
        - las bibliotecas son obligatorias para muchas tareas comunes, como entrada y salida
        
    - **comentarios**
        - el compilador de C++ ignora los comentarios
        - los comentarios son para que los programadores expliquen el proceso de pensamiento, bloques de código sutiles
        - Es una buena práctica escribir notas adecuadas como comentarios, especialmente. al aprender
        - facilita la lectura y comprensión del código sin tener que ejecutarlo y decodificarlo
        - *escribe código para que otros lo lean*
        - `//`: se utilizan barras diagonales dobles para comentarios de una sola línea
        - `/*` todo lo que hay dentro son comentarios; utilizado para comentarios de varias líneas `*/`
        
    - **códigos de instrucciones**
        - ¡le dice a la computadora qué hacer!
        - código compuesto por palabras clave, identificadores, símbolos, valores literales, etc.
            - cuando se combinan, seguir la gramática del idioma resuelve el problema
        - el bloque de códigos aparece entre llaves onduladas `{ }`
        - las declaraciones terminan con punto y coma (`;`)
        
    - **espacios en blanco**
        - El compilador normalmente ignora las sangrías, los espacios adicionales y las líneas en blanco a menos que sea necesario.
        - Se requieren espacios en blanco adecuados como mejor práctica para la legibilidad del código.

## Pruebas y depuración
- Los programas suelen contener muchos tipos de errores llamados bugs.
- Los programadores dedican la mayor parte de su tiempo a probar sus programas, encontrar errores y eliminarlos.
- el proceso de encontrar y corregir errores se llama depuración
- Muchos IDE proporcionan una forma de recorrer el código y examinar la memoria a medida que se ejecuta el programa.
- la clave para encontrar y corregir errores es probar 
    - Las pruebas exhaustivas garantizan que el programa proporcione resultados correctos para todos los conjuntos de entradas.

### Tipos de errores
- Hay tres tipos principales de errores: sintaxis, tiempo de ejecución y semánticos.

#### Errores en tiempo de compilación
- también llamados errores de sintaxis o errores gramaticales
- Los lenguajes informáticos son lenguajes formales con una gramática estricta de punto y coma.
    - Los lenguajes naturales (inglés, por ejemplo) están llenos de ambigüedad, redundancia y literalidad (modismos y metáforas).
- el compilador analiza el código C++; proporciona una lista de errores, si los hay.
- no puede compilar un programa en código de bytes si el programa tiene un error de compilación

#### Errores de tiempo de ejecución
- también llamadas excepciones de tiempo de ejecución
- estos errores aparecen mientras el programa se está ejecutando
- se puede manejar hasta cierto punto

#### Errores semánticos
- también llamados errores lógicos
    - errores en el proceso de pensamiento, pueden surgir debido a una mala comprensión del problema, solución/respuesta incorrecta, peculiaridades del lenguaje
- el programa funciona bien pero da una respuesta incorrecta
    - por ejemplo, sumar en lugar de multiplicar para resolver una ecuación (por ejemplo, $2+2$ es lo mismo que $2\times2$)
- puede identificarse y eliminarse realizando pruebas adecuadas
