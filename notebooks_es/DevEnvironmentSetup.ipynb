## Usando el compilador g++ en Windows WSL, Mac y Linux
- Los pasos proporcionados aquí suponen que estás utilizando el entorno de desarrollo C++ recomendado anteriormente.
- abrir un programa de Terminal
- estar familiarizado con la terminal y algunos [comandos bash básicos](https://sites.tufts.edu/cbi/files/2013/01/linux_cheat_sheet.pdf)
- cambie el directorio de trabajo actual a la carpeta correcta donde está el archivo .cpp
    - use el comando `ls` para ver todo el contenido del directorio
    - use el comando `cd <dir_name>` para cambiar el directorio al dir_name dado
- asegúrese de que el directorio de trabajo actual sea donde se encuentra su archivo `.cpp`
    - use el comando `pwd` en una terminal `*nix` para conocer el directorio de trabajo actual
- compilar usando g++
- ejecutar el ejecutable
- vale la pena recordar la siguiente secuencia de comandos
    - También puede usar estos comandos en repl.it IDE basado en la nube

```golpecito
$ cd project_folder # cambia el directorio de trabajo a la carpeta del proyecto
$ pwd # imprime el directorio de trabajo actual
$ ls # listar el contenido del directorio actual
$ g++ -std=c++17 -o programa_salida archivo_entrada.cpp # compilar archivo_entrada.cpp en programa_salida
$ ./output_program # ejecutar el programa de salida
```
