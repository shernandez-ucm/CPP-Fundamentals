## Entorno de aprendizaje de Jupyter Notebook

- Jupyter Notebook (herramienta popular en ciencia de datos y aprendizaje automático) es simplemente un entorno de aprendizaje.
    - no es un entorno de desarrollo profesional, especialmente. para programas C++
- Con Jupyter Notebook puedes aprender interactivamente muchos lenguajes de programación como Python, C, Rust, NodeJS, Bash, etc.
- le permite ejecutar código de forma interactiva, tomar notas con texto y HTML e incrustar archivos multimedia (imagen, audio, video, etc.)
- Lo más importante es que se pueden ejecutar códigos línea por línea y guardar el resultado.
    - Puede concentrarse en la línea de código para dominar realmente todos los detalles y conceptos básicos.
- También puedes leer y ejecutar estos cuadernos desde VS Code con las extensiones adecuadas
- La versión PDF de estos cuadernos se proporciona en la carpeta PDF del repositorio de GitHub.
    - los archivos PDF son de solo lectura; no puedo escribir y/o ejecutar código incrustado en archivos PDF

## Configurar el entorno de Jupyter Notebook
- se recomienda instalar el programa Jupyter Notebook en su computadora personal
- También necesitarás el kernel xeus-cling C++ para interpretar el código C++ y ver los resultados.
- siga las instrucciones de [https://github.com/rambasnet/DevEnvSetup](https://github.com/rambasnet/DevEnvSetup) para configurar Jupyter Notebook en varias plataformas
