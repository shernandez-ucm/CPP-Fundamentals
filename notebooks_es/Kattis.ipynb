### Descargar datos y metadatos de muestra del problema
- necesita una identificación del problema para descargar las muestras y los metadatos del problema
    - La identificación del problema se puede encontrar en la URL de la página del problema, p. para https://open.kattis.com/problems/carrots, la identificación del problema es `carrots`
- kattis-cli creará una carpeta con el nombre de identificación del problema y descargará todos los archivos de entrada y salida de muestra junto con los metadatos del problema dentro de esa carpeta.
- pasos para descargar y probar los problemas de Kattis usando kattis-cli:

```golpecito
cd <ruta a su directorio de trabajo>
es
kattis obtiene <problem_id>
es
cd <id_problema>
es
toque <id_problema>.cpp
# crea tu solución en el archivo .cpp anterior usando tu editor favorito
información de kattis
prueba de kattis
```

### Descarga y prueba manual de problemas de Kattis
- descargue las muestras proporcionadas en un archivo comprimido `.zip`
- descomprimir el archivo; creará una carpeta con el mismo nombre que el nombre del problema o el nombre del archivo zip
- cree un archivo de solución `problem_id.cpp` dentro de la misma carpeta donde están los archivos de muestra
- luego siga los siguientes pasos:
    - abrir una terminal en Mac/Linux/WSL
    - cambiar el directorio de trabajo a una carpeta con problemas, p.e. zanahorias

        ```golpecito
        persona con discapacidad
        cd <ruta a la carpeta de zanahorias>
        es
        ```

    - compilar directamente usando g++ o crear y usar un Makefile

        ```golpecito
        g++ -std=c++17 zanahorias.cpp
        ```

    - ejecutar casos de prueba de muestra proporcionados por Kattis, p. si 1.in y 1.ans son archivos de prueba de muestra correspondientes
    - lea el ejemplo 1.in y transmítalo al programa `a.out` y transfiera la respuesta del programa a diff para compararla con 1.ans

        ```golpecito
        gato 1.pulg | ./a.out | diferencia - 1.ans
        gato 2.pulgadas | ./a.out | diferencia - 2.ans
        ```
    - si la respuesta del programa es correcta, no verá ninguna diferencia ni salida en el terminal
    
- Además, puedes agregar estos comandos a tu Makefile
    - crea una regla, digamos kattis
    
    ```golpecito
    kattis:
        gato 1.pulg | ./a.out | diferencia - 1.ans
        gato 2.pulgadas | ./a.out | diferencia - 2.ans
        @echo "¡Todos los casos de prueba pasaron!"
    ```
    - ejecutar la regla desde la Terminal
    
    ```golpecito
    hacer kattis
    ```
    
- Una vez que su programa proporcione el resultado correcto como se muestra en el resultado correspondiente, cargue su archivo fuente `.cpp` en Kattis para probarlo con todas las muestras de prueba ocultas.

- Kattis compilará y ejecutará su programa para probarlo con otras muestras.
- Kattis aceptará su solución o la rechazará con algunos comentarios simples, como una respuesta incorrecta.
